```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler, MinMaxScaler, MaxAbsScaler, RobustScaler, Normalizer
```

```python
# Инициализация классификатора KNN
classifier = KNeighborsClassifier(n_neighbors=2)
```

```python
# Загружаем данные Iris и выбираем два признака
iris = load_iris()
iris_data = iris.data
X_train, X_test, y_train, y_test = train_test_split(iris_data, iris.target, test_size=0.3, random_state=0)
```

```python
# Список методов преобразования
scalers = {
    "Стандартный масштабатор": StandardScaler(),
    "Масштабатор Минимум-Максимум": MinMaxScaler(),
    "Масштабатор по максимальному абсолютному значению": MaxAbsScaler(),
    "Устойчивый масштабатор": RobustScaler(),
    "Нормализатор": Normalizer()
}

# Таблица для сравнения точности
results = []
```

```python
# Цикл по методам преобразования
for scaler_name, scaler in scalers.items():
    # Преобразование обучающих и тестовых данных
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)

    # Обучение и предсказание
    classifier.fit(X_train_scaled, y_train)
    y_pred = classifier.predict(X_test_scaled)

    # Оценка точности
    accuracy = accuracy_score(y_test, y_pred)

    # Сохраняем результаты
    results.append([scaler_name, accuracy])
```

```python
# Преобразуем результаты в DataFrame для удобства отображения
results_df = pd.DataFrame(results, columns=["Метод", "Точность"])
print(results_df)
```
